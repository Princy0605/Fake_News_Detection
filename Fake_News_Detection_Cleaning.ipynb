# Cleaning the Data

## Handling Null Values

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["text"])
```

```python
df["title"]=df["title"].fillna("")
```

```python
df.isnull().sum()
```

Missing values in the dataset were addressed by removing rows where the text field was null, as the main news content is essential for analysis, while null values in the title field were replaced with empty strings to preserve useful article information and enable effective combination of title and text without reducing the overall dataset size.

## Handling Duplicate Values

```python
df.duplicated().sum()
```

We see that there are no duplicates. Hence our dataset is free from any recurring values
